```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=500,noise=0.5,n_features=2,n_informative=2,random_state=42)
```

```python
print(X.shape , y.shape)
```

```python
plt.scatter(X[:,0],y)
plt.show()
```

```python
plt.scatter(X[:,1],y)
plt.show()
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
reg
```

```python
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X,y)
```
